```python
# FUNCIÓN CORREGIDA para trabajar con datos separados por punto y coma
def traducir_tipos_tui_corregida(valor):
    """
    Traduce los tipos separados por punto y coma
    Input: "fast_food_restaurant; restaurant; food; point_of_interest; establishment"
    Output: "Restaurante De Comida Rápida; Restaurantes; Comida; Lugar De Interés; Establecimiento"
    """
    if pd.isna(valor):
        return valor
    
    # Convertir a string y separar por punto y coma
    tipos = str(valor).split(';')
    tipos_traducidos = []
    
    for tipo in tipos:
        tipo_limpio = tipo.strip().lower()  # quitar espacios y convertir a minúsculas
        if tipo_limpio:  # si no está vacío
            traducido = map_en2es.get(tipo_limpio, f'__REVISAR__{tipo_limpio}')
            tipos_traducidos.append(traducido)
    
    # Volver a unir con punto y coma
    return '; '.join(tipos_traducidos)

# APLICAR LA TRADUCCIÓN
google['TIPOS_TUI'] = google['TIPOS_TUI'].apply(traducir_tipos_tui_corregida)

# VERIFICAR si quedó algo sin traducir
pendientes = google['TIPOS_TUI'].fillna('').str.contains('__REVISAR__', regex=False).sum()
print(f'Términos pendientes de traducir: {pendientes}')

# Ver algunos ejemplos de traducción
print("\nEjemplos de traducción:")
print(google['TIPOS_TUI'].head().tolist())
```

```python
import pandas as pd
import re, unicodedata
from collections import defaultdict
import numpy as np
import ast, re, codecs

# --- Normalización sin singularizar ---
def normalize_txt(s: str) -> str:
    s = str(s).strip().lower().replace("-", " ")
    s = "".join(c for c in unicodedata.normalize("NFKD", s) if not unicodedata.combining(c))
    s = re.sub(r"\s+", " ", s)
    return s

vowels = set("aeiou")

# Genera variantes razonables (singular/plural) SIN tocar el texto original
def variants(term: str):
    t = normalize_txt(term)
    vs = {t}
    if not t:
        return vs
    # plural desde singular
    if t[-1] in vowels:
        vs.add(t + "s")        # cine -> cines, restaurante -> restaurantes
    else:
        if t.endswith("z"):    # luz -> luces
            vs.add(t[:-1] + "ces")
        else:
            vs.add(t + "es")   # hospital -> hospitales
    # singular(es) desde plural: añadimos ambas posibilidades por seguridad
    if t.endswith("ces"):      # luces -> luz
        vs.add(t[:-3] + "z")
    if t.endswith("es"):
        if len(t) > 2:
            vs.add(t[:-2])     # hospitales -> hospital
        if len(t) > 1:
            vs.add(t[:-1])     # restaurantes -> restaurante; cines -> cine
    if t.endswith("s") and len(t) > 1:
        vs.add(t[:-1])         # bares -> bare (no se usará) pero también cubre cines -> cine
    return {x for x in vs if x}

# --- Tus grupos (añado Convento/Conventos) ---
grupos = {
    "TRANSPORTE": [
        "Estaciones De Autobús", "Estaciones De Metro", "Estaciones De Tren",
        "Estación De Transporte", "Alquiler De Coches"
    ],
    "GASOLINERAS Y APARCAMIENTOS": [
        "Aparcamientos", "Gasolineras", "Estación De Carga De Vehículos Eléctricos", "Lavadero De Coches",
        "Aparcamiento Disuasorio", "Área De Servicio Para Camiones"
    ],
    "ALOJAMIENTOS": [
        "Alojamiento", "B&B (Cama Y Desayuno)", "Cabañas", "Campings",
        "Casa De Huéspedes", "Edificio De Apartamentos",
        "Habitación De Huéspedes Privada", "Hostal", "Hotel De Larga Estancia",
        "Hotel Resort", "Hoteles", "Posada", "Área Para Autocaravanas", "Moteles",
        "Complejo De Apartamentos", "Complejo De Condominios", "Parque De Casas Móviles"
    ],
    "ESPACIOS DEPORTIVOS": [
        "Campos De Golf", "Centro De Fitness", "Club Deportivo", "Coaching Deportivo",
        "Complejo Deportivo", "Estudio De Yoga", "Gimnasios", "Instalación Deportiva",
        "Piscinas", "Saunas", "Zona De Senderismo",
        "Campo Deportivo", "Centro De Deportes De Aventura", "Pista De Hielo",
        "Parque De Ciclismo", "Estación De Esquí", "Parque De Skate", "Estanque De Pesca"
    ],
    "OCIO NOCTURNO": [
        "Discotecas", "Pub"
    ],
    "ADMINISTRACIÓN PÚBLICA Y SEGURIDAD": [
        "Ayuntamientos", "Oficina De Gobierno Local", "Oficina Gubernamental", "Policía", "Marina", "Embajada", "Subunidad"
    ],
    "COMIDA Y BEBIDA": [
        "Asador", "Bar De Vinos", "Bar Y Parrilla", "Bares", "Bocatería", "Cafetería Americana",
        "Cafés", "Café Para Perros", "Chocolatería", "Comida", "Comida Para Llevar",
        "Hamburguesería", "Heladerías", "Reparto De Comida", "Restaurante Africano",
        "Restaurante Asiático", "Restaurante Brasileño", "Restaurante Bufé", "Restaurante Chino",
        "Restaurante Coreano", "Restaurante De Alta Cocina", "Restaurante De Barbacoa",
        "Restaurante De Brunch", "Restaurante De Comida Rápida", "Restaurante De Desayunos",
        "Restaurante De Oriente Medio", "Restaurante De Ramen", "Restaurante De Sushi",
        "Restaurante Español", "Restaurante Estadounidense", "Restaurante Francés",
        "Restaurante Griego", "Restaurante Indio", "Restaurante Italiano", "Restaurante Japonés",
        "Restaurante Libanés", "Restaurante Mediterráneo", "Restaurante Mexicano",
        "Restaurante Tailandés", "Restaurante Turco", "Restaurante Vegano",
        "Restaurante Vegetariano", "Restaurante Vietnamita", "Restaurantes",
        "Panaderías", "Pastelerías", "Patio De Comidas", "Pizzería", "Marisquerías", "Salón De Té",
        "Cibercafé", "Restaurante De Postres"
    ],
    "ESPACIOS RELIGIOSOS": [
        "Iglesias", "Lugar De Culto", "Mezquitas", "Cementerio", "Templo Hindú"
    ],
    "TIENDAS DE COMIDA": [
        "Carnicerías", "Charcuterías", "Mercados", "Mayorista", "Supermercado Asiático", "Supermercados"
    ],
    "TIENDAS": [
        "Tienda De Alimentación", "Tienda De Artículos Para El Hogar", "Tienda De Açaí",
        "Tienda De Bagels", "Tienda De Bicicletas", "Tienda De Bricolaje",
        "Tienda De Comestibles", "Tienda De Conveniencia", "Tienda De Deportes",
        "Tienda De Descuentos", "Tienda De Donuts", "Tienda De Dulces", "Tienda De Electrónica",
        "Tienda De Licores", "Tienda De Mascotas", "Tienda De Muebles", "Tienda De Móviles",
        "Tienda De Regalos", "Tienda De Repuestos", "Tienda De Ropa", "Tienda De Zumos",
        "Tiendas", "Zapaterías", "Gran Almacén", "Floristerías", "Confitería",
        "Ferretería", "Librerías", "Joyerías", "Centros Comerciales", "Tienda Tipo Almacén"
    ],
    "MODA Y BELLEZA": [
        "Barbería", "Cuidado Del Cabello", "Esteticista", "Estudio De Tatuajes Y Piercings",
        "Peluquerías", "Salón De Belleza", "Salón De Uñas", "Spas", "Sastre", "Maquillador", "Cuidado De Pies", "Centro De Bronceado"
    ],
    "EDUCACIÓN": [
        "Bibliotecas", "Colegios", "Escuela Primaria", "Instituto", "Universidades", "Preescolar"
    ],
    "SALUD Y MEDICINA": [
        "Salud", "Clínica Dental", "Clínica Dermatológica", "Dentistas", "Droguerías",
        "Farmacias", "Hospitales", "Fisioterapeuta", "Masajes", "Laboratorio Médico",
        "Médico", "Quiropráctico", "Centro De Bienestar", "Veterinaria"
    ],
    "CULTURA, ARTE Y NATURALEZA": [
        "Centro Cultural", "Lugar De Interés", "Escultura", "Monumento",
        "Museos", "Plaza", "Sitio Histórico", "Teatro De Artes Escénicas", 'Lugar Emblemático', 
        "Galería De Arte", "Estudio De Arte", "Acuario", "Jardín", "Jardín Botánico", "Zona De Picnic",
        "Granjas", "Zoológicos", "Parque Nacional", "Lugar Histórico", "Centro De Información Turística", "Mirador", "Planetario"
    ],
    "OCIO": [
        "Cine", "Club De Comedia", "Centro De Ocio",
        "Boleras", "Parque Infantil",
        "Parques", "Parques De Atracciones", "Salón Recreativo",
        "Atracción Turística", "Campamento Infantil", "Casino", "Karaoke"
    ],
    "EVENTOS Y CELEBRACIONES": [
        "Sala De Baile", "Sala De Conciertos", "Sala De Eventos", "Salón De Bodas", "Recinto",
        "Servicio De Catering", "Auditorio", "Salón De Banquetes", "Estadio", "Pabellón", "Establecimiento", "Centro De Convenciones", "Centro Comunitario"
    ],
    "SERVICIOS": [
        "Cerrajero", "Electricista", "Fontanero", "Empresa De Mudanzas", "Empresa De Pintura",
        "Inmobiliarias", "Agencia De Cuidado Infantil", "Agencia De Excursiones", "Agencia De Viajes",
        "Almacenaje", "Aseguradora", "Consultoras", "Contratista General", "Mensajería",
        "Oficina Corporativa", "Operador De Telecomunicaciones", "Organizador De Campamentos De Verano",
        "Servicios Financieros", "Tanatorios", "Taller Mecánico", "Concesionario",
        "Abogado", "Alquiler De Películas", "Aseos Públicos", "Banco", "Cajero Automático", "Contabilidad",
        "Lavandería", "Contratista De Techos", "Oficina De Correos", "Dirección"
    ]
}



# --- Diccionario tipo_normalizado -> set(categorías) con variantes ---
tipo_a_categorias = defaultdict(set)
for cat, items in grupos.items():
    for it in items:
        for v in variants(it):
            tipo_a_categorias[v].add(cat)

# --- Conversión de una celda "A, B, C" -> "CAT1, CAT2, ..." ---
def categorias_para_celda(cell) -> str:
    if pd.isna(cell) or not str(cell).strip():
        return "OTROS"
    cats = set()
    for raw in str(cell).split(","):
        key = normalize_txt(raw)
        if key in tipo_a_categorias:
            cats.update(tipo_a_categorias[key])
    return ", ".join(sorted(cats)) if cats else "OTROS"

# --- Aplicar ---
google["CATEGORIA_TUI"] = google["TIPOS_TUI"].apply(categorias_para_celda)

# --- Diagnóstico de no mapeados (ya NO verás 'cin' ni 'restaurant') ---
def tipos_no_mapeados(series):
    acc = set()
    for cell in series.dropna().astype(str):
        for x in cell.split(","):
            x = normalize_txt(x)
            if x and x not in tipo_a_categorias:
                acc.add(x)
    return sorted(acc)

no_map = tipos_no_mapeados(google["TIPOS_TUI"])
print("No mapeados (muestra):", no_map[:30])

google['RESERVA_POSIBLE'] = google['RESERVA_POSIBLE'].map({'VERDADERO': 'SI', 'FALSO': 'NO'})
google['ACCESIBILIDAD_SILLA_RUEDAS'] = google['ACCESIBILIDAD_SILLA_RUEDAS'].map({'VERDADERO': 'SI', 'FALSO': 'NO'})

google['RATING_TUI'] = google['RATING_TUI']/10
google['PRECIO'] = google['PRECIO'].fillna('—')
google['ESTADO_NEGOCIO'] = google['ESTADO_NEGOCIO'].fillna('—')
google['URL'] = google['URL'].fillna('—')

google = google.reset_index(drop=True)
google['PLACE_ID'] = google.index + 1
google['PLACE_ID'] = google['PLACE_ID'].astype('int64')

################################ LIMPIAR HORARIO ############################
import re, ast, numpy as np

DAYS = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']

def _deescape_u(match):
    # \uXXXX -> carácter real (Unicode)
    return chr(int(match.group(1), 16))

def _deescape_x(match):
    # \xNN -> decodifica como cp1252 (p.ej., \x96 -> EN DASH U+2013)
    b = int(match.group(1), 16)
    return bytes([b]).decode('cp1252', errors='ignore')

def limpiar_token(t: str) -> str:
    # 1) Decodifica secuencias literales \uXXXX y \xNN
    t = re.sub(r'\\u([0-9a-fA-F]{4})', _deescape_u, t)
    t = re.sub(r'\\x([0-9a-fA-F]{2})', _deescape_x, t)

    # 2) Sustituye espacios invisibles por espacio normal y compacta
    t = (t.replace('\u2009', ' ')
           .replace('\u202f', ' ')
           .replace('\u00A0', ' ')
           .replace('\uFEFF', ' '))
    t = re.sub(r'\s+', ' ', t).strip()

    # 3) Fallback si quedó U+0096 (misdecode) -> EN DASH
    t = t.replace('\u0096', '–')

    # 4) Normaliza cualquier guion/raya a EN DASH y elimina espacios alrededor
    t = re.sub(r'\s*[–—−\-‒―]\s*', '–', t)

    # 5) AM/PM en mayúsculas y pegado al número
    t = re.sub(r'(?i)\b(am|pm)\b', lambda m: m.group(1).upper(), t)
    t = re.sub(r'\s+(AM|PM)\b', r'\1', t)

    # 6) Normaliza textos típicos
    t = re.sub(r'(?i)\bopen\s*24\s*hours\b', 'Open 24 hours', t)
    t = re.sub(r'(?i)\bclosed\b', 'Closed', t)

    return t

def parse_list_like(s: str):
    s = s.strip()
    # Caso típico: "['Monday: ...', 'Tuesday: ...', ...]"
    if s.startswith('[') and s.endswith(']'):
        try:
            lst = ast.literal_eval(s)
            if isinstance(lst, list):
                return [str(x) for x in lst]
        except Exception:
            pass
    # Si viene plano, partimos por | o comas delante de un Día:
    if any(d + ':' in s for d in DAYS):
        parts = re.split(r"'\s*,\s*'|\s*\|\s*|\s*,\s*(?=[A-Z][a-z]+:)", s)
        return [p.strip(" '\"") for p in parts if ':' in p]
    return None

def normalizar_horario_cell(cell):
    if pd.isna(cell):
        return np.nan

    t = str(cell)
    # Limpia caracteres/escapes
    t = limpiar_token(t)

    items = parse_list_like(t)
    if not items:
        # Si ya viene con todos los días "Day: ..." separados por |
        if all(d + ':' in t for d in DAYS):
            parts = [p.strip() for p in t.split('|')]
            m = {}
            for p in parts:
                if ':' in p:
                    d, v = p.split(':', 1)
                    m[d.strip()] = limpiar_token(v)
            return ' | '.join(f'{d}: {m.get(d, "Closed")}' for d in DAYS)
        return np.nan

    # Construye diccionario día->valor
    day2val = {}
    for it in items:
        it = limpiar_token(it)
        if ':' not in it:
            continue
        d, v = it.split(':', 1)
        d, v = d.strip(), limpiar_token(v)
        if d in DAYS:
            day2val[d] = v

    # Ensambla en orden fijo y rellena días faltantes con Closed
    return ' | '.join(f'{d}: {day2val.get(d, "Closed")}' for d in DAYS)

# === APLICAR ===
google['HORARIO'] = google['HORARIO'].apply(normalizar_horario_cell)

################################ LIMPIAR LATITUD Y LONGITUD ##############################
import re
import numpy as np
import pandas as pd

def to_float_simple(x):
    """
    Convierte string -> float sin NaN:
    - Quita comillas y espacios invisibles (u2009, u202f, NBSP, etc).
    - Cambia coma decimal por punto (p.ej., -3,742E+16 -> -3.742E+16).
    - Si falla, limpia caracteres no numéricos comunes y reintenta.
    - Si aún falla, devuelve 0.0 (para evitar vacíos).
    """
    if pd.isna(x):
        return 0.0
    s = str(x).strip()
    if len(s) >= 2 and ((s[0] == s[-1] == "'") or (s[0] == s[-1] == '"')):
        s = s[1:-1].strip()
    s = re.sub(r'[\u2000-\u200F\u202F\u2060\u00A0\uFEFF\s]+', '', s)  # espacios raros
    s = re.sub(r'(?<=\d),(?=\d)', '.', s)  # coma decimal -> punto

    try:
        return float(s)
    except Exception:
        s2 = re.sub(r'[^0-9eE+\-\.]', '', s)
        try:
            return float(s2)
        except Exception:
            return 0.0

def scale_lat(v):
    """
    Si LATITUD_TUI está en [10,100) no hace nada.
    Si está en [100,1000) divide entre 10,
    si en [1000,10000) divide entre 100, y así sucesivamente (solo dividir).
    """
    x = abs(float(v))
    while x >= 100:
        x /= 10.0
    return x

def scale_lon(v):
    """
    Para LONGITUD_TUI: si |lon| está en [0,10) no hace nada; si en [10,100) divide entre 10;
    si en [100,1000) divide entre 100; ... y siempre devuelve NEGATIVO.
    """
    x = abs(float(v))
    while x >= 10:
        x /= 10.0
    return -x

# 1) STRING -> NUMÉRICO (sin escalar)
lat_raw = google['LATITUD_TUI'].apply(to_float_simple).astype('float64')
lon_raw = google['LONGITUD_TUI'].apply(to_float_simple).astype('float64')

print("ANTES DE ESCALAR:")
print(pd.DataFrame({'LATITUD_TUI': lat_raw, 'LONGITUD_TUI': lon_raw}).describe())

# 2) ESCALADO POR MAGNITUD (solo dividir según rangos)
lat_scaled = lat_raw.apply(scale_lat).astype('float64')
lon_scaled = lon_raw.apply(scale_lon).astype('float64')

print("\nDESPUÉS DE ESCALAR:")
print(pd.DataFrame({'LATITUD_TUI': lat_scaled, 'LONGITUD_TUI': lon_scaled}).describe())

# 3) ESCRIBIR EN LAS COLUMNAS (sin crear nuevas columnas)
google['LATITUD_TUI']  = lat_scaled
google['LONGITUD_TUI'] = lon_scaled
```

```python
a_quitar = {"lugar de interés", "establecimiento"}

def limpiar_tipos(s):
    if pd.isna(s):
        return s
    s = str(s)
    # quita el prefijo "Tipos:" si existe y parte el resto por comas
    prefijo = "Tipos:"
    cuerpo = s.split(":", 1)[1] if ":" in s else s
    items = [x.strip() for x in cuerpo.split(",") if x.strip()]

    # filtra los que no quieres (comparación case-insensitive)
    filtrados = [it for it in items if it.casefold() not in a_quitar]

    # reconstruye con el prefijo
    return f"{prefijo} {', '.join(filtrados)}" if filtrados else f"{prefijo}"

# aplica a todo el DataFrame
google["TIPOS_TUI"] = google["TIPOS_TUI"].apply(limpiar_tipos)
```

```python
# --- Normalización de CATEGORIA_TUI a una sola categoría dominante (solo si hay dominancia) ---
import re
from collections import Counter, defaultdict
import pandas as pd

def _parse_multivalue(cell):
    """Divide cadenas tipo 'a, b | c / d; e' en lista normalizada."""
    if pd.isna(cell):
        return []
    tokens = re.split(r'[,\|/;]+', str(cell))
    return [t.strip() for t in tokens if t and t.strip()]

def _norm(s):
    return str(s).strip().upper()

def _invert_category_to_types(category_to_types):
    """Convierte {categoria: [tipos]} -> {tipo: [categororias]}."""
    tipo_to_cats = defaultdict(list)
    for cat, tipos in (category_to_types or {}).items():
        if tipos is None:
            continue
        if isinstance(tipos, str):
            tipos_list = _parse_multivalue(tipos)
        else:
            try:
                iter(tipos)
                tipos_list = list(tipos)
            except TypeError:
                tipos_list = [tipos]
        for tipo in tipos_list:
            tnorm = _norm(tipo)
            cnorm = _norm(cat)
            if cnorm not in tipo_to_cats[tnorm]:
                tipo_to_cats[tnorm].append(cnorm)
    return dict(tipo_to_cats)

def _ensure_tipo_to_cats():
    """
    Obtiene un mapeo {TIPO_TUI -> [CATEGORIAS]} intentando:
      1) usar TIPO_A_CATEGORIA si existe (tipo->categoria[s]),
      2) o invertir 'grupos' si existe (categoria->tipos),
      3) o un mínimo fallback (AMPLÍA según tu dominio).
    """
    # 1) tipo -> categorias
    try:
        mapping = TIPO_A_CATEGORIA
        out = defaultdict(list)
        for tipo, cats in mapping.items():
            t = _norm(tipo)
            if cats is None:
                continue
            if isinstance(cats, str):
                cats_list = _parse_multivalue(cats)
            else:
                try:
                    iter(cats)
                    cats_list = list(cats)
                except TypeError:
                    cats_list = [cats]
            for c in cats_list:
                cn = _norm(c)
                if cn not in out[t]:
                    out[t].append(cn)
        return dict(out)
    except NameError:
        pass

    # 2) categorias -> tipos (p.ej., 'grupos')
    try:
        return _invert_category_to_types(grupos)
    except NameError:
        pass

    # 3) fallback mínimo: completa con tus tipos reales
    return {
        _norm('Complejo Deportivo'): [_norm('ESPACIOS DEPORTIVOS')],
        _norm('Gimnasios'): [_norm('ESPACIOS DEPORTIVOS')],
        _norm('Instalación Deportiva'): [_norm('ESPACIOS DEPORTIVOS')],
    }

# Mapeo tipo -> [categorias]
TIPO_TO_CATS = _ensure_tipo_to_cats()

def _elegir_categoria_dominante(row):
    """Devuelve la categoría dominante si hay una única ganadora; en empate o sin mapeo, deja el valor original."""
    tipos = [_norm(t) for t in _parse_multivalue(row.get('TIPOS_TUI'))]
    cat_counts = Counter()
    for tipo in tipos:
        for c in TIPO_TO_CATS.get(tipo, []):
            cat_counts[c] += 1

    # Sin mapeo: no tocar
    if not cat_counts:
        return row.get('CATEGORIA_TUI')

    # Ganadora(s)
    max_count = max(cat_counts.values())
    candidatas = [c for c, n in cat_counts.items() if n == max_count]

    # Única ganadora -> cambiar
    if len(candidatas) == 1:
        return candidatas[0]

    # Empate -> dejar tal cual
    return row.get('CATEGORIA_TUI')

def normalizar_categoria_dominante(df_):
    """Aplica la regla a todo el DataFrame."""
    if 'CATEGORIA_TUI' not in df_.columns or 'TIPOS_TUI' not in df_.columns:
        raise ValueError("Faltan columnas requeridas: CATEGORIA_TUI y/o TIPOS_TUI")

    # Copia de auditoría (si no existe)
    if 'CATEGORIA_TUI_ORIG' not in df_.columns:
        df_['CATEGORIA_TUI_ORIG'] = df_['CATEGORIA_TUI']

    # Reescritura condicionada
    df_['CATEGORIA_TUI'] = df_.apply(_elegir_categoria_dominante, axis=1)
    return df_

google = normalizar_categoria_dominante(google)
```
